```python
from scipy import stats
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def calc_vif(df):
    vif=pd.DataFrame()
    vif["variables"]=df.columns
    vif["VIF FACTOR"]=[variance_inflation_factor(df.values,i) for i in range(df.shape[1])]
    return(vif)
```

```python
calc_vif(df)
```
